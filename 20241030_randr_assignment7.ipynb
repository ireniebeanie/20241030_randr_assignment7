Using repositories, tracking changes, and publishing can help with computational reproducibility because it allows you to eliminate unecessary factors of variability in data analysis. A lot of modern biology includes -omics experiments that generate a huge amount of data that can be analyzed. Depending on your research question, the way you analyze the data will be different. Furthermore, the way in which you analyze the data will lead you to different conclusions about the results of the data. If you decide to trim FASTQ sequences, but don't include that step in a repository, others who try and replicate your experimental results/analysis might get different results than you, decreasing the reproducibility of your study and making your results less convincing to the scientific community. As you continue to improve your code and method of analysis, it is important to track these changes so that, once again, others who want to do a similar experiment as you can follow the same steps that you did and can reliably draw conclusions from their experiment.
